# Feature Extraction (Color features)

```python
# Install required libraries if not already installed
!pip install opencv-python-headless tqdm

# Import required libraries
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Dataset and output folder (Update the path below to your dataset location in Drive)
# Dataset and output folder
dataset_path = "/content/Resize mango leaf disease"  #  Your updated dataset path
features_folder = "/content/Resize mango leaf disease/Features"
output_csv_name = "color_features.csv"
output_csv_path = os.path.join(features_folder, output_csv_name)

IMG_SIZE = 256  # Resize size

# Create Features folder if not exist
if not os.path.exists(features_folder):
    os.makedirs(features_folder)
    print(f"Created folder: {features_folder}")

features = []

# Feature Extraction Loop
for class_name in sorted(os.listdir(dataset_path)):
    class_folder_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder_path):
        print(f"\nExtracting features for class: {class_name}")

        for file_name in tqdm(os.listdir(class_folder_path), desc=f"Processing images in '{class_name}'"):
            file_path = os.path.join(class_folder_path, file_name)

            if os.path.isfile(file_path):
                try:
                    img = cv2.imread(file_path)
                    if img is None:
                        continue

                    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

                    # Split into channels
                    b, g, r = cv2.split(img)

                    # HSV conversion
                    hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
                    h, s, v = cv2.split(hsv)

                    # Means
                    mean_r = np.mean(r)
                    mean_g = np.mean(g)
                    mean_b = np.mean(b)

                    # Standard Deviations
                    std_r = np.std(r)
                    std_g = np.std(g)
                    std_b = np.std(b)

                    mean_h = np.mean(h)
                    mean_s = np.mean(s)
                    mean_v = np.mean(v)

                    # Green pixel percentage
                    lower_green = np.array([25, 40, 40])
                    upper_green = np.array([85, 255, 255])
                    green_mask = cv2.inRange(hsv, lower_green, upper_green)
                    green_pixel_percent = np.sum(green_mask > 0) / (IMG_SIZE * IMG_SIZE) * 100

                    # Flattened RGB Histogram (32 bins per channel)
                    hist_r = cv2.calcHist([r], [0], None, [32], [0, 256]).flatten()
                    hist_g = cv2.calcHist([g], [0], None, [32], [0, 256]).flatten()
                    hist_b = cv2.calcHist([b], [0], None, [32], [0, 256]).flatten()

                    # Normalize histograms
                    hist_r = hist_r / hist_r.sum()
                    hist_g = hist_g / hist_g.sum()
                    hist_b = hist_b / hist_b.sum()

                    # Combine all features
                    feature_row = [
                        mean_r, mean_g, mean_b,
                        std_r, std_g, std_b,
                        mean_h, mean_s, mean_v,
                        green_pixel_percent
                    ]

                    feature_row.extend(hist_r)
                    feature_row.extend(hist_g)
                    feature_row.extend(hist_b)
                    feature_row.append(class_name)

                    features.append(feature_row)

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Define column names
columns = [
    'mean_red', 'mean_green', 'mean_blue',
    'std_red', 'std_green', 'std_blue',
    'mean_hue', 'mean_saturation', 'mean_value',
    'green_pixel_percentage'
]

columns += [f'hist_r_{i}' for i in range(32)]
columns += [f'hist_g_{i}' for i in range(32)]
columns += [f'hist_b_{i}' for i in range(32)]
columns.append('class_label')

# Save to CSV
df = pd.DataFrame(features, columns=columns)
df.to_csv(output_csv_path, index=False)
print(f"\nFeature extraction completed and saved to: {output_csv_path}")
```

# Shape features

```python
# Install required packages
!pip install opencv-python-headless tqdm

# Import required libraries
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm

# Dataset and output folder ( Update the path below if needed)
dataset_path = "/content/Resize mango leaf disease"
features_folder = os.path.join(dataset_path, "Features")
output_csv_name = "shape_features.csv"
output_csv_path = os.path.join(features_folder, output_csv_name)

IMG_SIZE = 256  # Resize size

# Create Features folder if not exist
if not os.path.exists(features_folder):
    os.makedirs(features_folder)
    print(f"Created folder: {features_folder}")

shape_features = []

# Shape Feature Extraction Loop
for class_name in sorted(os.listdir(dataset_path)):
    class_folder_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder_path):
        print(f"\nExtracting shape features for class: {class_name}")

        for file_name in tqdm(os.listdir(class_folder_path), desc=f"Processing {class_name}"):
            file_path = os.path.join(class_folder_path, file_name)

            if os.path.isfile(file_path):
                try:
                    # Read image in grayscale
                    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                    if img is None:
                        continue

                    # Resize image
                    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

                    # Threshold the image to get binary image
                    _, thresh = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

                    # Find contours
                    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                    if not contours:
                        continue
                    contour = max(contours, key=cv2.contourArea)

                    # Calculate shape features
                    area = cv2.contourArea(contour)
                    perimeter = cv2.arcLength(contour, True)
                    aspect_ratio = 0
                    circularity = 0

                    # Aspect ratio (width / height of bounding box)
                    x, y, w, h = cv2.boundingRect(contour)
                    if h != 0:
                        aspect_ratio = float(w) / h

                    # Circularity (using perimeter and area)
                    if perimeter != 0:
                        circularity = (4 * np.pi * area) / (perimeter ** 2)

                    # Count contour points
                    contour_points_count = len(contour)

                    # Append the features for this image
                    shape_features.append([
                        area,
                        perimeter,
                        aspect_ratio,
                        contour_points_count,
                        circularity,
                        class_name
                    ])

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Define column names
columns = [
    'area', 'perimeter', 'aspect_ratio',
    'contour_points_count', 'circularity', 'class_label'
]

# Save shape features to CSV
df = pd.DataFrame(shape_features, columns=columns)
df.to_csv(output_csv_path, index=False)
print(f"\n Shape feature extraction completed and saved to: {output_csv_path}")
```

# Texture Features

```python
# Install required libraries (if not already installed)
!pip install opencv-python-headless scikit-image pywavelets tqdm

# Import necessary libraries
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm
import pywt
from skimage.feature import graycomatrix, graycoprops, local_binary_pattern
from skimage.filters import gabor
from skimage import img_as_ubyte
from scipy.stats import entropy

# Dataset and output folder ( Update if using different path)
dataset_path = "/content/Resize mango leaf disease"
features_folder = os.path.join(dataset_path, "Features")
output_csv_name = "texture_features.csv"
output_csv_path = os.path.join(features_folder, output_csv_name)

IMG_SIZE = 256  # Resize size
LBP_RADIUS = 1
LBP_POINTS = 8 * LBP_RADIUS
LBP_METHOD = 'uniform'
GLCM_DISTANCES = [1]
GLCM_ANGLES = [0]

# Create Features folder if not exists
if not os.path.exists(features_folder):
    os.makedirs(features_folder)
    print(f"Created folder: {features_folder}")

# List to hold extracted texture features
texture_features = []

# Texture Feature Extraction Loop
for class_name in sorted(os.listdir(dataset_path)):
    class_folder_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder_path):
        print(f"\nExtracting texture features for class: {class_name}")

        for file_name in tqdm(os.listdir(class_folder_path), desc=f"Processing {class_name}"):
            file_path = os.path.join(class_folder_path, file_name)

            if os.path.isfile(file_path):
                try:
                    # Read image in grayscale
                    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                    if img is None:
                        continue

                    # Resize image
                    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
                    img_uint8 = img_as_ubyte(img)

                    # GLCM features
                    glcm = graycomatrix(img_uint8, distances=GLCM_DISTANCES, angles=GLCM_ANGLES, levels=256, symmetric=True, normed=True)
                    glcm_contrast = graycoprops(glcm, 'contrast')[0, 0]
                    glcm_correlation = graycoprops(glcm, 'correlation')[0, 0]
                    glcm_homogeneity = graycoprops(glcm, 'homogeneity')[0, 0]

                    # Entropy of the image histogram
                    hist = cv2.calcHist([img_uint8], [0], None, [256], [0, 256])
                    hist = hist.ravel() / hist.sum()
                    texture_entropy = entropy(hist, base=2)

                    # LBP (Local Binary Pattern)
                    lbp = local_binary_pattern(img_uint8, P=LBP_POINTS, R=LBP_RADIUS, method=LBP_METHOD)
                    (hist_lbp, _) = np.histogram(lbp.ravel(), bins=np.arange(0, LBP_POINTS + 3), range=(0, LBP_POINTS + 2))
                    hist_lbp = hist_lbp.astype("float")
                    hist_lbp /= (hist_lbp.sum() + 1e-6)
                    lbp_histogram_mean = hist_lbp.mean()

                    # Gabor Filter response
                    gabor_response, _ = gabor(img_uint8, frequency=0.6)
                    gabor_mean_response = gabor_response.mean()

                    # DWT (Discrete Wavelet Transform) Energy
                    coeffs = pywt.dwt2(img_uint8, 'db1')
                    _, (cH, cV, cD) = coeffs
                    dwt_energy = np.sum(cH ** 2) + np.sum(cV ** 2) + np.sum(cD ** 2)

                    # Append extracted features for the current image
                    texture_features.append([
                        glcm_contrast,
                        glcm_correlation,
                        glcm_homogeneity,
                        texture_entropy,
                        lbp_histogram_mean,
                        gabor_mean_response,
                        dwt_energy,
                        class_name
                    ])

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Define column names for the CSV file
columns = [
    'glcm_contrast', 'glcm_correlation', 'glcm_homogeneity',
    'texture_entropy', 'lbp_histogram_mean', 'gabor_mean_response',
    'dwt_texture_energy', 'class_label'
]

# Save the texture features to CSV
df = pd.DataFrame(texture_features, columns=columns)
df.to_csv(output_csv_path, index=False)
print(f"\n Texture feature extraction completed and saved to: {output_csv_path}")
```

# Edge Feature

```python
# Install required libraries (if not already installed)
!pip install opencv-python-headless tqdm

# Import necessary libraries
import os
import cv2
import numpy as np
import pandas as pd
from tqdm import tqdm

# Dataset and output folder ( Update if using different path)
dataset_path = "/content/Resize mango leaf disease"  # Update this path for Colab
features_folder = os.path.join(dataset_path, "Features")
output_csv_name = "edge_features.csv"
output_csv_path = os.path.join(features_folder, output_csv_name)

IMG_SIZE = 256  # Resize size

# Create Features folder if not exists
if not os.path.exists(features_folder):
    os.makedirs(features_folder)
    print(f"Created folder: {features_folder}")

# List to hold extracted edge features
edge_features = []

# Edge Feature Extraction Loop
for class_name in sorted(os.listdir(dataset_path)):
    class_folder_path = os.path.join(dataset_path, class_name)

    if os.path.isdir(class_folder_path):
        print(f"\nExtracting edge features for class: {class_name}")

        for file_name in tqdm(os.listdir(class_folder_path), desc=f"Processing {class_name}"):
            file_path = os.path.join(class_folder_path, file_name)

            if os.path.isfile(file_path):
                try:
                    # Read image in grayscale
                    img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
                    if img is None:
                        continue

                    # Resize image
                    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

                    # Canny Edge Detection
                    canny_edges = cv2.Canny(img, 100, 200)
                    canny_edge_count = np.sum(canny_edges > 0)
                    edge_pixel_ratio = canny_edge_count / (IMG_SIZE * IMG_SIZE)

                    # Sobel Gradient Magnitude
                    sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
                    sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
                    sobel_gradient = np.sqrt(sobel_x**2 + sobel_y**2)
                    sobel_gradient_mean = np.mean(sobel_gradient)
                    sobel_gradient_std = np.std(sobel_gradient)

                    # Laplacian Variance
                    laplacian = cv2.Laplacian(img, cv2.CV_64F)
                    laplacian_variance = laplacian.var()

                    # Append extracted edge features
                    edge_features.append([
                        canny_edge_count,
                        edge_pixel_ratio,
                        sobel_gradient_mean,
                        sobel_gradient_std,
                        laplacian_variance,
                        class_name
                    ])

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

# Define column names for the CSV file
columns = [
    'canny_edge_count', 'edge_pixel_ratio',
    'sobel_gradient_mean', 'sobel_gradient_std',
    'laplacian_variance', 'class_label'
]

# Save the edge features to CSV
df = pd.DataFrame(edge_features, columns=columns)
df.to_csv(output_csv_path, index=False)
print(f"\n Edge feature extraction completed and saved to: {output_csv_path}")
```

```python
# Install required libraries
!pip install -q scikit-learn matplotlib seaborn

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# File path to the shape features
file_path = '/content/Resize mango leaf disease/Features/shape_features.csv'

# Load the CSV
df = pd.read_csv(file_path)

# Check for missing values
if df.isnull().sum().any():
    print("Warning: Missing values found. Filling with column mean.")
    df.fillna(df.mean(), inplace=True)

# Display first few rows
print("Data Preview:")
print(df.head())

# Separate features and labels
X = df.drop(columns=['class_label'])
y = df['class_label']

# Scale features (important for SVM)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset with stratification
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Initialize the SVM classifier (RBF kernel)
svm_classifier = SVC(kernel='rbf', random_state=42)

# Train the model
svm_classifier.fit(X_train, y_train)

# Predict on the test set
y_pred = svm_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Install required libraries
!pip install -q scikit-learn matplotlib seaborn

# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# File path to edge features
file_path = '/content/Resize mango leaf disease/Features/edge_features.csv'

# Load the CSV
df = pd.read_csv(file_path)

# Preview the data
print("Data Preview:")
print(df.head())

# Check for missing values
if df.isnull().sum().any():
    print("Missing values found. Filling with column means.")
    df.fillna(df.mean(), inplace=True)

# Separate features and labels
X = df.drop(columns=['class_label'])
y = df['class_label']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split the dataset (stratified to preserve label distribution)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, stratify=y, random_state=42
)

# Initialize KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors=5)  # Try 3, 5, 7, 9 to tune

# Train the model
knn_classifier.fit(X_train, y_train)

# Predict on test set
y_pred = knn_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.4f}")

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)

# Plot the Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()

# Classification Report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Merging
import os
import pandas as pd

# Paths to individual feature CSV files
base_path = "/content/Resize mango leaf disease/Features"
color_features_path = os.path.join(base_path, "color_features.csv")
shape_features_path = os.path.join(base_path, "shape_features.csv")
texture_features_path = os.path.join(base_path, "texture_features.csv")
edge_features_path = os.path.join(base_path, "edge_features.csv")

# Load feature CSVs
color_df = pd.read_csv(color_features_path)
shape_df = pd.read_csv(shape_features_path)
texture_df = pd.read_csv(texture_features_path)
edge_df = pd.read_csv(edge_features_path)

# Extract class label (assumed consistent across all CSVs)
class_label = color_df['class_label']

# Drop class_label from feature sets to avoid duplication
color_df = color_df.drop(columns=['class_label'], errors='ignore')
shape_df = shape_df.drop(columns=['class_label'], errors='ignore')
texture_df = texture_df.drop(columns=['class_label'], errors='ignore')
edge_df = edge_df.drop(columns=['class_label'], errors='ignore')

# Combine all feature DataFrames
features_df = pd.concat([color_df, shape_df, texture_df, edge_df], axis=1)

# Add the class label back
features_df['class_label'] = class_label

# Save to a new CSV
output_path = os.path.join(base_path, "features.csv")
features_df.to_csv(output_path, index=False)

print(f"\n Combined feature file saved to: {output_path}")
```

```python
# Scaling and combining csv features
# Install required library if not already installed
!pip install -q scikit-learn

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the combined feature dataset
file_path = "/content/Resize mango leaf disease/Features/features.csv"  # Adjust path as needed
df = pd.read_csv(file_path)

# Separate features and labels
X = df.drop(columns=['class_label'])
y = df['class_label']

# Split into train and test sets (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardize the feature values (mean=0, std=1)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Output the shape of the resulting datasets
print(" Feature scaling complete.\n")
print(f"X_train shape: {X_train_scaled.shape}")
print(f"X_test shape: {X_test_scaled.shape}")
```

```python
# SVM
# Install required libraries if not already installed
!pip install -q scikit-learn matplotlib seaborn

# Import additional libraries for plotting
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

# Create and train the SVM model
from sklearn.svm import SVC
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train_scaled, y_train)

# Make predictions
svm_y_pred = svm_model.predict(X_test_scaled)

# Accuracy
svm_accuracy = accuracy_score(y_test, svm_y_pred)
print(f" SVM Model Accuracy: {svm_accuracy:.4f}\n")

# Classification report
print(" Classification Report:\n")
print(classification_report(y_test, svm_y_pred))

# Confusion matrix
cm = confusion_matrix(y_test, svm_y_pred)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.title(" Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
# Random Forest
# Install required libraries (only needed once in Colab)
!pip install scikit-learn matplotlib seaborn

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load dataset (use your path)
df = pd.read_csv('/content/Resize mango leaf disease/Features/features.csv')

# Separate features and target
X = df.drop(columns=['class_label'])
y = df['class_label']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train Random Forest model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train_scaled, y_train)

# Predict on the test set
rf_y_pred = rf_model.predict(X_test_scaled)

# Accuracy & Report
print(f" Random Forest Accuracy: {accuracy_score(y_test, rf_y_pred):.4f}\n")
print(" Classification Report (Random Forest):\n")
print(classification_report(y_test, rf_y_pred))

# Confusion Matrix
plt.figure(figsize=(8,6))
sns.heatmap(confusion_matrix(y_test, rf_y_pred), annot=True, fmt='d', cmap='Oranges',
            xticklabels=sorted(y.unique()), yticklabels=sorted(y.unique()))
plt.title("Random Forest - Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
